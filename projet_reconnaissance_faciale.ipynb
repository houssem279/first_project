```python
import cv2,os # fournit des fonctionnalités pour le traitement d'images via la bibliothèque OpenCV
import numpy as np #pour effectuer des calculs numériques,
```

```python
data_path='projectai/CK+48' #chemin vers le répertoire contenant les données des images.

data_list= os.listdir(data_path) #stocker les liste dans le repertoire dans data liste

img_data=[]  #Initialise une liste vide pour stocker les données des images après traitement.
#charge les images dans chaque dossier, les redimensionne en 48x48 pixels et les ajoute à la liste img_data
for dataset in data_list:
    img_list= os.listdir(data_path+'/'+ dataset)
    print('Loading the images from the folder -{}\n'.format(dataset))
    for img in img_list:
        input_img=cv2.imread(data_path+'/'+dataset+'/'+img) 
        input_img_resized= cv2.resize(input_img,(48,48))
        img_data.append(input_img_resized)
    #convertit une liste d'images en un tableau numérique, ajuste le type de données et normalise les valeurs des pixels, puis affiche sa forme.    
data=np.array(img_data)
data= data.astype('float32')
data= data/255.0
data.shape
```

```python
from sklearn.utils  import shuffle #Cette fonction est utilisée pour mélanger aléatoirement les données.
from keras.utils import to_categorical #Cette fonction est utilisée pour convertir des vecteurs d'étiquettes 
from sklearn.model_selection import train_test_split  # diviser les données en ensembles d'entraînement et de test.

# Détermine le nombre de classes en comptant le nombre de dossiers (classes) dans data_list
num_classes = len(data_list)
num_samples = data.shape[0]

# Assignation des étiquettes à chaque classe en fonction des plages d'indices 
labels = np.ones((num_samples,), dtype='int64')

labels[0:135] = 0 #135 (anger)
labels[135:189] = 1 #54 (comtempt)
labels[189:366] = 2 #177 (disgust)
labels[366:441] = 3 #75(fear)
labels[441:648] = 4 #207 (happy)
labels[648:732] = 5  #84 (Sadness)
labels[732:981] = 6 #249 (suprise)



# Associe les noms des classes (noms de dossiers) à la variable names
names = data_list
img_labels = to_categorical(labels, num_classes)

# Convertit les étiquettes en encodage one-hot adapté à l'utilisation dans un modèle d'apprentissage automatique
x_img,y_img = shuffle(data,img_labels,random_state=2)

# Utilise la fonction train_test_split pour diviser les données en ensembles d'entraînement et de test
x_train,x_test,y_train,y_test = train_test_split (x_img,y_img,test_size=0.2,random_state=2)
```

```python
import matplotlib.pyplot as plt  # Importe le module pyplot de la bibliothèque Matplotlib

# Crée une nouvelle figure de taille 20x20 pouces pour afficher les images
plt.figure(figsize=(20,20))

for i,j in enumerate(list(np.random.randint (0,len(x_img),36))): # une liste de 36 indices aléatoires dans la plage de 0 à la longueur de x_img
    plt.subplot(6,6,i+1) # Crée un sous-tracé dans une grille de 6x6 à la position i+1 dans la figure
    plt.imshow(x_img[j])
    plt.axis('off')
    class_index = np.argmax(y_img[j])
    plt.title(names[class_index])# Utilise le nom de classe correspondant à l'index pour attribuer un titre à l'image
         
 


        
        
```

# Construire le modèle CNN

```python

from keras.models import Sequential # utilisée pour créer des modèles de réseaux de neurones séquentiels.
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D


def create_model():
    input_shape = (48, 48, 3)  # Définit la forme de l'entrée comme une image de 48x48 pixels 
    
    model = Sequential()  # Initialise un modèle séquentiel

    # Ajoute une couche de convolution avec 6 filtres de taille 5x5, activation ReLU, et même taille de sortie
    model.add(Conv2D(6, (5,5), padding='same', activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2,2)))  # Ajoute une couche de mise en commun avec un pool de 2x2

    # Ajoute une autre couche de convolution avec 16 filtres de taille 5x5, activation ReLU, et même taille de sortie
    model.add(Conv2D(16, (5,5), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))  # Ajoute une autre couche de mise en commun

    model.add(Flatten())  # Aplati les données en vue de les connecter à une couche dense

    model.add(Dense(64, activation='relu'))  # Ajoute une couche dense avec 64 neurones et activation ReLU

    model.add(Dense(128, activation='relu'))  # Ajoute une autre couche dense avec 128 neurones et activation ReLU
    model.add(Dropout(0.5))  
    model.add(Dense(7, activation='softmax'))  # Ajoute une couche dense de sortie avec une activation softmax pour les 7 classes possibles
    # Compile le modèle 
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])  
    return model  # Renvoie le modèle construit
```

```python
#crée un modèle de réseau de neurones 
```

```python
model= create_model()
model.summary()
```

```python
# Entraînement d'un modèle CNN pour la reconnaissance d'émotions sur des images
```

```python
history = model.fit(x_train, y_train, epochs=200, validation_data=(x_test, y_test))
```

```python
#testing 
```

```python
loss, score = model.evaluate(x_test, y_test, verbose=0) #valuation le model
# Affiche le score de précision pour l'ensemble de test, converti en pourcentage
print('the acccuarcy for tesring :%f' % (score * 100))
# Fonction pour prédire et afficher une image avec sa classe prédite
def predict_sample(index):
    y_predict = model.predict(x_test[index:index+1])
    

    plt.imshow(x_test[index]) # Affiche l'image de l'échantillon testé 
    plt.title(names[list(y_predict.astype(int).flatten()).index(1)])# Affiche le titre de l'image prédite en utilisant le nom correspondant à la classe prédite

    
predict_sample(77)

print(names[list(y_test[77]).index(1)])
# Affiche le nom correspondant à la vraie classe de l'échantillon testé (index 1) dans le dataset

```

```python
#effectuer la comparaison et tracer les valeurs de pertes
```

```python
train_loss = history.history['loss']# Récupère les valeurs de perte (loss) 

val_loss = history.history['val_loss']# Récupère les valeurs de perte (loss) s

train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']# Récupère validation


steps = range(len(train_loss))# Crée une liste de valeurs représentant les époques d'entraînement (nombre d'étapes/époques)


plt.plot(steps, train_loss, 'r', label='train_loss')
# Trace la courbe de la perte (loss) sur l'ensemble d'entraînement en fonction du nombre d'époques
plt.plot(steps, val_loss, 'g', label='val_loss')
# Trace la courbe de la perte (loss) sur l'ensemble de validation en fonction du nombre d'époques
plt.title('train_loss vs val_loss')# Ajoute un titre au graphique
plt.legend()# Ajoute une légende au graphique pour différencier train_loss et val_loss
plt.figure()# Crée une nouvelle figure pour un autre graphique


plt.plot(steps, train_accuracy, 'r', label='train_accuracy')
# Trace la courbe de la précision sur l'ensemble d'entraînement en fonction du nombre d'époques
plt.plot(steps, val_accuracy, 'g', label='val_accuracy')
plt.title('train_accuracy vs val_accuracy')
plt.legend()
plt.figure()
```
